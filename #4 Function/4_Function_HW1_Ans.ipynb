# 1  迴文判斷

建立一個函式用以判斷輸入的字串是否是迴文(反轉字串跟原本一樣)，<br>
並使用迴圈呼叫函式執行至少三次判斷結果。

範例:
```
Enter a number for loop: 3
akfm3
akfm3 is not a palindrome.
dk3kd
dk3kd is a palindrome.
lvmkmvl
lvmkmvl is a palindrome.
```

````python
def palindrome(inp_str):

    if inp_str == inp_str[::-1]:
        return True
    else:
        return False

count = int(input("Enter a number for loop: "))

for _ in range(count):

    inp_str = input()
    if palindrome(inp_str):
        print(inp_str+" is a palindrome.")
    else:
        print(inp_str+" is not a palindrome.")
````

# 2 字串平方

建立一個函式用以將輸入字串中的數字正序排序後，以數字為key對應到value為平方的方式建成字典回傳，再印出此字典結果。<br>
**Hint**: 使用list comprehension, dict comprehension完成此題、判斷字串中的數字用`string.isdigit()`。

範例:
```
Enter a string: 9d4j2
{2: 4, 4: 16, 9: 81}
Enter a string: 209rjf35
{0: 0, 2: 4, 3: 9, 5: 25, 9: 81}
Enter a string: 18c0e302
{0: 0, 1: 1, 2: 4, 3: 9, 8: 64}
```

````python
def str_pow(num_str):
    num_ls = [int(s) for s in list(num_str) if s.isdigit()]
    num_ls.sort()
    num_dict = {n:n**2 for n in num_ls}
    return num_dict

inp = input("Enter a string: ")
while inp!='q':
    inp_dict = str_pow(inp)
    print(inp_dict)
    inp = input("Enter a string: ")
````

# 3 羅馬數字轉換

設計一個函式可以將輸入的數字轉換成羅馬數字表示。

|Arabic|Roman||Arabic|Roman||Arabic|Roman||Arabic|Roman|
|---|---|---|---|---|---|---|---|---|---|---|
|1|I||4|IV||5|V||9|IX|
|10|X||40|XL||50|L||90|XC|
|100|C||400|CD||500|D||900|CM|
|1000|M|

規則: [羅馬數字](https://en.wikipedia.org/wiki/Roman_numerals)<br>
**Hint**: 將上表建成字典作轉換

範例:
```
Enter a number: 123
CXXIII
Enter a number: 48
XLVIII
Enter a number: 1991
MCMXCI
Enter a number: 8
VIII
```

````python
romanDict = {"M":1000, "CM":900, "D":500, "CD":400, "C":100, 
              "XC":90, "L":50, "XL":40, "X":10, "IX":9, "V":5, "IV":4, "I":1}

def int2roman(num):
    result=""
    for roman, value in romanDict.items():
        while num >= value:
            num -= value
            result += roman
    return result

inp = input("Enter a number: ")
while inp!='q':
    romanNum = int2roman(int(inp))
    print(romanNum)
    inp = input("Enter a number: ")
````
