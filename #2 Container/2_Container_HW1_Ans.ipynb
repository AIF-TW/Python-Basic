# 1 list.dict操作

使用`input()`輸入一序列 由個位數字與字母交錯且不重複(以逗號區分) 的字串，<br>
將此輸入字串依逗號進行切分後，分別存成只有數字的`num_ls`和只有字母的`alpha_ls`，並印出這兩個list。<br>
建構一個字典將`alpha_ls`進行正序排序後做為字典的key，而`num_ls`進行反向排序後作為相對位置的value，並印出此字典。


範例:
```python
3,g,6,s,9,n,2,d
['3', '6', '9', '2']
['g', 's', 'n', 'd']
{'d': '9', 'g': '6', 'n': '3', 's': '2'}
```


````python
inp_str = input()
str_ls = inp_str.split(',')

num_ls = str_ls[::2]
print(num_ls)
num_ls.sort(reverse=True)

alpha_ls = str_ls[1::2]
print(alpha_ls)
alpha_ls.sort()

dictionary = dict(zip(alpha_ls, num_ls))
print(dictionary)
````

# 2 水果價目表

建構一個水果價目表的字典(如下圖)，<br>
水果品項作為字典的key，對應的value以list形式存放兩個值分別為單價和盒裝價。

|fruit|$|box|
|---|---|---|
|apple|30|180|
|banana|10|60|
|grape|80|250|
|peach|60|500|
|orange|20|120|


完成以下敘述並印出結果:
1. 印出此字典
2. 計算買了 兩顆蘋果 和 三條香蕉 的價錢
3. 蘋果單價漲價至45、桃子盒裝價降價至480，請更新此價目表
4. 假設手上有$300的預算，請計算出**兩種**符合此預算且至少購入一盒水果的方式
5. 葡萄下架了且另外上架了梨子(pear)，單價與盒裝價分別30,200，請更新此價目表


````python
fruit_dict = {'apple':[30, 180],
              'banana':[10, 60],
              'grape':[80, 250],
              'peach':[60, 500],
              'orange':[20,120]}

print(fruit_dict)

price2 = fruit_dict['apple'][0]*2+fruit_dict['banana'][0]*3
print(price2)

fruit_dict['apple'][0] = 45
fruit_dict['peach'][1] = 480
print(fruit_dict)

price3 = fruit_dict['grape'][1]+fruit_dict['orange'][0]*2+fruit_dict['banana'][0]
print(price3)

price4 = fruit_dict['banana'][1]+fruit_dict['orange'][1]*2
print(price4)

fruit_dict.pop('grape')
fruit_dict['pear'] = [30, 200]
print(fruit_dict)
````
