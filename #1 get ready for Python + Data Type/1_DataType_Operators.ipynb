#### format 格式化

改變字串格式的方法

#### 1: %-formatting

透過`%`格式符號，將()中的變數依照指定格式方式輸出<br>

|format|code|
|---|---|
|整數　　　　　|   %d   |
|八進制整數　　|   %o   |
|十六進制整數　|   %x   |
|浮點數　　　　|   %f   |
|科學記號   |   %e   |
|字串　　　　　|   %s   |

````python
aPrice = 50
bPrice = 15

# 在要印變數的地方輸入變數的 format code，並在字串後面輸入 % 後擺放變數
print("An apple cost %s dollars." % (aPrice))
print("An apple cost %d dollars and a banana cost %d." % (aPrice, bPrice))
````

````python
pi = 3.141592653589793

# format code 也能用於指定輸出位數格式
print("Pi = %f " % (pi))
print("Pi = %.2f " % (pi)) # 指定小數點位數
print("Pi = %8.2f " % (pi)) # 指定總共位數&小數點位數
print("Pi = %08.2f " % (pi)) # 以0填空至指定位數
````

`{}`中以`:` 轉換格式

````python
var_int = 12345

print(f'num = {var_int}')   # 正常輸出變數
print(f'num = {var_int:d}')  # 格式化十進制d/二進制b/八進制o/十六進制x
print(f'num = {var_int:,}')  # 逗號區隔
print(f'num = {var_int:6d}')  # 格式化指定(6)位數
print(f'num = {var_int:06d}')  # 格式化指定(6)位數，並以0填空
````

````python
var_float = 0.12345

print(f'num = {var_float}')
print(f'num = {var_float:%}')   # 百分比表示
print(f'num = {var_float:.4f}')  # 指定小數點位數
print(f'num = {var_float:8.4f}')  # 指定總共位數&小數點位數
print(f'num = {var_float:08.4f}') # 以0填空至指定位
````

````python
var_str = "abc"

# 自定義置左及置中，以及填補空位的字元

print(f'str = {var_str:>15s}')  # 佔指定格數(15)置右，其餘以空白填補
print(f'str = {var_str:*^12s}')  # 佔指定格數(12)置中，其餘以*填補
print(f'str = {var_str:~<10s}') # 佔指定格數(10)置左，其餘以*填補
````

## Arithmetic 算數


算數運算子與數值資料進行一般數學運算

| Operator | Description |
|----|---|
| +  | Addition |
| -  | Subtraction |
| *  | Multiplication |
| /  | Division |
| %  | Modulus |
| **  | Exponentiation |
| //  | Floor division |

````python
x = 10; y = 3

print(f'{x} + {y} = {x + y}')
print(f'{x} - {y} = {x - y}')
print(f'{x} * {y} = {x * y}')
print(f'{x} / {y} = {x / y}')

# 取餘數
print(f'{x} % {y} = {x % y}')

# 冪次寫法，x 的 y 次冪
print(f'{x} ** {y} = {x ** y}')

# 整除，只留下商
print(f'{x} // {y} = {x // y}')
````

## Comparison 比較

比較兩個值，回傳boolean值(`True` or `False`)

| Operator | Description |
|----|---|
| == | Equal |
| !=  | Not equal |
| < | Less than |
| > | Greater than |
| <=  | Less than or equal to |
| >=  | Greater than or equal to |

````python
# 數值比較
a = 5; b = 6.5

print(f'{a} == {b}? {a == b}')
print(f'{a} != {b}? {a != b}')
print(f'{a} < {b}? {a < b}')
print(f'{a} > {b}? {a > b}')
print(f'{a} <= {b}? {a <= b}')
print(f'{a} >= {b}? {a >= b}')
````

````python
# 字串比較
a = 'abc'; b = 'abd'

# 字串大小比較是依照字典順序，先出現者較小。
print(f'{a} == {b}? {a == b}')
print(f'{a} != {b}? {a != b}')
print(f'{a}  > {b}? {a > b}')
print(f'{a}  < {b}? {a < b}')
print(f'{a} >= {b}? {a >= b}')
print(f'{a} <= {b}? {a <= b}')
````

## Logical 邏輯

結合條件判斷

|Operator|Description|
|---|---|
|and|Returns True if `both` statements are true|
|or|Returns True if `one of the` statements is true|
|not|Reverse the result|

````python
x = 5; y = 6
print(x > 1 and y <10)
print(x > 6 or y <10)
print(not x > 1 )
````
