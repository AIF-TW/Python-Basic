#  1 矩形計算機

使用`input()`輸入兩個正整數，分別表示矩形的**長**和**寬**，計算此矩陣的**周長**、**面積**與**對角線**，並使用`print()`以兩位小數的精準度印出。

範例:
```
長 : 3
寬 : 4
周長 : 14.00
面積 : 12.00
對角線 : 5.00
```

````python
import math
x, y = float(input("長 : ")), float(input("寬 : "))
print("周長 : %.2f"%(2*(x+y)))
print("面積 : %.2f"%(x*y))
print("對角線 : %.2f"%(math.sqrt(x**2+y**2)))
````

# 2 時間轉換器

使用`input()`輸入秒數，將秒數轉換成`__ 日 __ 時 __ 分 __ 秒`的形式，並使用`print()`印出。

範例:
```
輸入秒數 : 2468080
28 日 13 時 34 分 40 秒
```

````python
time = int(input("輸入秒數 : "))
day = int(time / 60 / 60 / 24)
time = time - (day * 60 * 60 * 24)
hour = int(time / 60 / 60)
time = time - (hour * 60 * 60)
min = int(time / 60)
time = time - (min * 60)
sec = time

print(day, "日", hour, "時", min, "分", sec, "秒")
````

# 3 字串切分

```python
numStr = '1a2b3c4d5e6f'
```
使用slicing語法，將上述字串切分成以下結果印出。
```
123456
abcdef
654321
fedcba
```

````python
numStr = '1a2b3c4d5e6f'

print(numStr[::2])
print(numStr[1::2])

print(numStr[-2::-2])
print(numStr[::-2])
````
